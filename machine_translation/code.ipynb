```python
flores_rus = load_dataset('facebook/flores', 'rus_Cyrl')
flores_eng = load_dataset('facebook/flores', 'eng_Latn')
```

```python
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-600M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M")
```

```python
individual_bleu_scores = []

for translation, reference in zip(nllb_eng_rus_mydataset, mydataset_rus_formatted):
    # Note: sacrebleu expects a list of references for each translation, hence [reference]
    bleu_score = sacrebleu.sentence_bleu(translation, reference).score
    individual_bleu_scores.append(bleu_score)

# Now, individual_bleu_scores contains the BLEU score for each sentence
print(len(individual_bleu_scores))

min_value = min(individual_bleu_scores)
max_value = max(individual_bleu_scores)

# Find the indices of the minimum and maximum values
min_index = individual_bleu_scores.index(min_value)
max_index = individual_bleu_scores.index(max_value)

print("Minimum value:", min_value, "at index", min_index)
print("Maximum value:", max_value, "at index", max_index)
print(nllb_eng_rus_mydataset[min_index])
print(my_dataset_rus[min_index])
print(nllb_eng_rus_mydataset[max_index])
print(my_dataset_rus[max_index])
```
